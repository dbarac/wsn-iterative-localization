### Testiranje na slučajnim mrežama
Algoritam je testiran na 10 slučajno generiranih mreža sa 20 čvorova sa mjerenjima bez šuma. Slike sa crvenim
čvorovima prikazuju stvarni položaj čvorova. Slike sa plavim čvorovima prikazuju lokalizirane čvorove 
u relativnom koordinatnom sustavu.

```python
import scipy.stats
import numpy as np
from pymote import (
    NetworkGenerator, Node, Network, write_pickle, Simulation
)
from pymote.sensor import DistSensor
from pymote.algorithms.localization import (
    IterativeLocalization,
    network_is_rigid,
    network_is_generically_globaly_rigid,
)

trueDistSensor = DistSensor(
    {'pf': scipy.stats.norm, 'scale': 0} # no measurement noise
)

net_gen = NetworkGenerator(
    20, sensors=('NeighborsSensor', trueDistSensor), comm_range=350
)

n_tested_networks = 0
while n_tested_networks < 10:
    Node.cid = 1
    net = net_gen.generate_random_network()

    if not (network_is_rigid(net) and 
            network_is_generically_globaly_rigid(net)):
        print("Discarding network ^")
        continue

    net.show()
    net.algorithms = (IterativeLocalization, )

    write_pickle(
        net, "random-localization-net{}.npc.gz".format(n_tested_networks)
    )

    sim = Simulation(net)
    sim.run()

    for n in net.nodes():
        print(n.id, n.status)

    # show localized nodes in relative coordinate system
    # (defined by initiator node)
    net.show(
        positions={n: np.array(n.memory['position']) for n in net.nodes()},
        nodeColor='blue'
    )
    n_tested_networks += 1
```
